# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides=ride_data_df.groupby('city').count()

total_rides[['ride_id']]
```

```python
# 2. Get the total drivers for each city type

total_drivers=city_data_df.groupby('city').sum('driver_count')
total_drivers
```

```python
#  3. Get the total amount of fares for each city type
total_amount_fares=ride_data_df.groupby('city').sum('fare')

total_amount_fares[['fare']]
```

```python
#  4. Get the average fare per ride for each city type.




frames=pd.concat([total_amount_fares[['fare']],total_rides[['ride_id']]],axis=1)
frames['AVERAGE_FARE_PER_RIDE']=frames['fare']/frames['ride_id']
frames
```

```python
# 5. Get the average fare per driver for each city type. 
frames=pd.concat([total_drivers,total_amount_fares[['fare']]],axis=1)
frames['AVERAGE_FARE_PER_DRIVER']=frames['fare']/frames['driver_count']
frames
```

```python
#  6. Create a PyBer summary DataFrame. 
frame_total_ride=pyber_data_df.groupby('type').count()[['ride_id']]

frame_total_fare=pyber_data_df.groupby('type').sum('fare')[['fare']]

frame_total_driver=city_data_df.groupby('type').sum('driver_count')[['driver_count']]

titles=['Total Rides','Total Drivers','Total Fares','Average Fare per Ride','Average Fare per Driver']
pyber_summary_df=pd.concat([frame_total_ride,frame_total_driver,frame_total_fare],axis=1)
pyber_summary_df['Average Fare per Ride']=pyber_summary_df['fare']/pyber_summary_df['ride_id']
pyber_summary_df['Average Fare per Driver']=pyber_summary_df['fare']/pyber_summary_df['driver_count']
pyber_summary_df.columns=titles

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pd.options.display.float_format = '${:,.2f}'.format
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
pyber_data_df.groupby(['city','date']).sum('fare')[['fare']]
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
pyber_data_df.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

pyber_pivot=pyber_data_df.pivot(index='date',columns='type')

pyber_pivot['fare']
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

pyber_pivot.loc['2019-01-01':'2019-04-28']['fare']
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

pyber_pivot.index=pd.to_datetime(pyber_pivot.index)

```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_pivot.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_resample_weekly=pyber_pivot['fare'].resample('W').sum()
pyber_resample_weekly
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
plt.style.use('fivethirtyeight')

plt.figure(figsize=(20,5))
plt.plot(pyber_resample_weekly.index,pyber_resample_weekly['Rural'],linewidth=2,label='Rural')
plt.plot(pyber_resample_weekly.index,pyber_resample_weekly['Suburban'],linewidth=2,label='Suburban')
plt.plot(pyber_resample_weekly.index,pyber_resample_weekly['Urban'],linewidth=2,label='Urban')
plt.title('Total Fare by City Type')

plt.ylabel('Fare($USD)')
plt.savefig('analysis/PyBer_fare_summary.png')
plt.legend(title='type',loc='center')



plt.show()


```
